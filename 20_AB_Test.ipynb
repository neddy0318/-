# A/B Test

## 1. 개요

&nbsp;&nbsp; - A/B 테스트란 기존 서비스(A)와 새로 적용하고자하는 서비스(B), 혹은 그에 따른 유저 집단을 A, B로 나눠 통계적으로 비교하는 것입니다. 이를 통해 새로운 서비스가 기존 서비스에 비해 효과적인지를 정량적으로 평가하는 방법입니다.

&nbsp;&nbsp; - 과학에서 수행하는 무작위비교연구(RCT; Randomized-Controlled Trial)를 마케팅에 적용한 것입니다. 주로 인터넷 마케팅에 많이 사용하나, 인터페이스, 디자인, 상품 배치 등을 개선하기 위해서도 사용합니다.

## 2. 왜 하는가?

&nbsp;&nbsp; - 그렇다면 왜 A/B 테스트를 수행하는 걸까요? 그 이유는 바로 상관관계로부터 인과관계(정확하게는 '인과관계일 확률이 높은 것')를 찾아내고자 함입니다.

&nbsp;&nbsp; - 이를 통해 '원인'이 되는 것에 개입함으로써 실제 결과를 원하는 방향으로 바꾸거나 개선하는 것이 A/B 테스트의 목표입니다.

&nbsp;&nbsp; - 가령 새로운 광고 도입이 정말로 매출 증대로 이어졌는지, 혹은 다른 요인 때문인 것은 아닌지를 확인하는 작업을 수행하는 것이 A/B 테스트를 하는 이유입니다.

## 3. 단계

&nbsp;&nbsp; -A/B 테스트는 다음의 단계로 이루어집니다.

&nbsp;&nbsp; 1)가설 설정 

&nbsp;&nbsp;&nbsp; - A/B 테스트에서 가설은 검증하고자 하는 것이 무엇인지를 과학적인 방법(통계적 방법)에 기반해 참/거짓을 판별할 수 있는 문장 형태로 구체화하는 것입니다.

&nbsp;&nbsp;&nbsp; - ex)'광고를 B 형태로 바꾸면 기존 A 형태보다 더 많은 이들이 구매할 것이다.'

&nbsp;&nbsp; 2)메트릭 정의

&nbsp;&nbsp;&nbsp; - 가설이 참인지 거짓인지를 뒷받침할 수 있는 측정기준을 메트릭(Metric)이라고 합니다.

&nbsp;&nbsp;&nbsp; - 메트릭은 하나의 가설에 대해 둘 이상일 수 있습니다. 위의 경우 '구매 횟수'와 '구매 취소 횟수' 두 가지가 있을 수 있습니다.

&nbsp;&nbsp; 3)실험 설계

&nbsp;&nbsp;&nbsp; - 실험 설계를 진행할 때는 집단을 둘로 나눠 하나는 기존 환경(A)에, 다른 하나는 실험 환경(B)에 배정합니다.

&nbsp;&nbsp;&nbsp; - 이 때 실험 환경(B)은 정확한 해석을 위해 검증할 요소 외 다른 요소들은 기존 환경과 모두 동일해야 합니다. 또한 집단 배정 시에도 무작위 배정(Random Assingment)을 해줘야합니다.

&nbsp;&nbsp; 4)결과 도출

&nbsp;&nbsp;&nbsp; - 실험 설계를 완료한 후 두 집단에 대해 미리 정의한 메트릭을 계산합니다(위의 경우 '구매 횟수'를 계산합니다). 

&nbsp;&nbsp;&nbsp; - 그리고 통계적으로 유의미한 차이가 있는지, 그에 따라 B가 A보다 나은지 판단합니다. 주로

## 5. 결과 신뢰성 검증

&nbsp;&nbsp; **1)AA Test**

&nbsp;&nbsp;&nbsp; - AA 테스트는 A/B 테스트의 신뢰성을 검증하기 위해 사용하는 방법입니다. A/B 테스트는 두 가지 다른 버전을 비교하는 반면, AA 테스트는 동일한 두 버전을 비교합니다. 이 테스트를 통해 시스템 자체가 오류를 일으키거나, 실험 방법론에 문제가 있는지를 확인할 수 있습니다.

&nbsp;&nbsp;&nbsp; -AA 테스트의 목적:

&nbsp;&nbsp;&nbsp;&nbsp; (1)랜덤화 검증: 실험 참가자들이 무작위로 배정되었는지 확인합니다.

&nbsp;&nbsp;&nbsp;&nbsp; (2)기저선 데이터 확인: 실험 조건이 동일할 때 결과가 유사한지 확인합니다.

&nbsp;&nbsp;&nbsp;&nbsp; (3)시스템 검증: 실험 플랫폼이나 데이터 수집 시스템이 올바르게 작동하는지 확인합니다.

&nbsp;&nbsp; **2)P-Value**

&nbsp;&nbsp;&nbsp; - p-value는 통계적 검정에서 관찰된 결과가 귀무가설(null hypothesis) 하에서 나타날 확률을 나타냅니다. A/B 테스트에서 p-value는 두 버전 간의 차이가 우연에 의한 것인지 아니면 실제로 유의미한 차이가 있는지를 판단하는 데 사용됩니다.

&nbsp;&nbsp;&nbsp; - p-value의 해석:

&nbsp;&nbsp;&nbsp;&nbsp;작은 p-value (보통 0.05 이하): 귀무가설을 기각할 수 있는 증거가 강력합니다. 즉, 두 버전 간의 차이가 통계적으로 유의미하다고 판단합니다.

&nbsp;&nbsp;&nbsp;&nbsp;큰 p-value (보통 0.05 이상): 귀무가설을 기각할 수 있는 증거가 부족합니다. 즉, 두 버전 간의 차이가 우연에 의한 것일 가능성이 큽니다.

&nbsp;&nbsp;&nbsp; - AA 테스트와 p-value의 관계 : AA 테스트에서도 p-value를 사용할 수 있습니다. 두 동일한 버전을 비교했을 때, p-value가 0.05 이하로 나오면 이는 문제를 시사합니다. 왜냐하면 동일한 두 버전이 통계적으로 유의미한 차이를 보일 이유가 없기 때문입니다. 따라서, 이런 경우 실험 설계나 데이터 수집 과정에 문제가 있을 가능성을 의심해야 합니다.

There seems to be a mathematical difference between the two groups. But is this difference a chance occurrence or is it statistically significant?

1. Establish the hypotheses:
H0: M1 = M2 (There is no statistically significant difference between the mean ages of those with and without diabetes)
H1: M1! = M2 (There is a statistically significant difference between the mean ages of those with and without diabetes)
2. Examine Assumptions
Assumption of normality

H0: The assumption of normal distribution is satisfied.
H1: Assumption of normal distribution is not satisfied

```python
# H0 REJ if p-value < 0.05.
# H0 CANNOT BE REJECTED if p-value < 0.05.
```

```python
test_stat, pvalue = shapiro(df.loc[df["Outcome"] == 1, "Age"].dropna())
```

```python
print('Test Stat = %.4f, p-value = %.4f' % (test_stat, pvalue))
```

```python
test_stat, pvalue = shapiro(df.loc[df["Outcome"] == 0, "Age"].dropna())
```

```python
print('Test Stat = %.4f, p-value = %.4f' % (test_stat, pvalue))
```

```python
# Since the p-value is less than 0.05, hypothesis H0 is rejected.
# The assumption of normality is not met.
```

```python
# Hipotez (H0: M1 = M2)
```

```python
test_stat, pvalue = mannwhitneyu(df.loc[df["Outcome"] == 1, "Age"].dropna(),
                                 df.loc[df["Outcome"] == 0, "Age"].dropna())
```

```python
print('Test Stat = %.4f, p-value = %.4f' % (test_stat, pvalue))
```

```python
# H0 hypothesis is rejected.
# The hypothesis that there is no statistically significant difference between the mean ages of people with and without diabetes is rejected.
# There is a statistically significant difference between the mean ages of those with and without diabetes.

# In other words, those with higher age have diabetes.
```

```python
def get_bootstrap_samples(data, n_samples=1000):
    indices = np.random.randint(0, len(data), (n_samples, len(data)))
    samples = data[indices]
    return samples

def stat_intervals(stat, alpha=0.05):
    boundaries = np.percentile(stat, [100 * alpha / 2., 100 * (1 - alpha / 2.)])
    return boundaries
```

```python
control = get_bootstrap_samples(df.loc[df.VARIANT_NAME == 'control', 'REVENUE'].values, 10000)
variant = get_bootstrap_samples(df.loc[df.VARIANT_NAME == 'variant', 'REVENUE'].values, 10000)
```

```python
def plot_distribution_and_stat_intervals(variant, control, title, alpha=0.05):
    """ Plot the distribution of the mean difference and return the confidence intervals """
    f, ax = plt.subplots()
    # get data for coloring confidence intervals
    points = sns.kdeplot(variant - control, shade=False).get_lines()[0].get_data()
    x = points[0]
    y = points[1]
    ymin, ymax = plt.ylim()
    # highlight the zero value and the bounds of the confidence interval
    plt.vlines(0, 0, ymax, label='0', color='gray')
    plt.vlines(stat_intervals(variant - control, alpha)[0], 0, ymax, linestyles="dashed")
    plt.vlines(stat_intervals(variant - control, alpha)[1], 0, ymax, linestyles="dashed")
    # color the confidence interval and zones outside it
    plt.fill_between(x,y,
                     where = (x >= stat_intervals(variant - control, alpha)[1]),
                     color='gainsboro')
    plt.fill_between(x,y,
                     where = (x <= stat_intervals(variant - control, alpha)[0]),
                     color='gainsboro')
    plt.fill_between(x,y,
                     where = ((x >= stat_intervals(variant - control, alpha)[0])
                              & (x <= stat_intervals(variant - control, alpha)[1])),
                     color='red',
                     label = '95% confidence interval')
    plt.title(f'Distribution of difference between means (variant - control) {title}; {100*(1-alpha)}% Confidence interval for difference of means: {stat_intervals(variant - control, alpha)}')
    plt.legend(prop={'size':13})
    # return confidence interval data
    return stat_intervals(variant - control)
```

```python
plot_distribution_and_stat_intervals(
    np.mean(variant, axis=1), 
    np.mean(control, axis=1),
    title='all users'
)

# f, ax = plt.subplots()
# get data for coloring confidence intervals
# points = sns.kdeplot(np.mean(variant, axis=1) - np.mean(control, axis=1), shade=True).get_lines()[0].get_data()
# np.mean(variant, axis=1) - np.mean(control, axis=1)
```
