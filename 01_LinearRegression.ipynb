# **1. 선형회귀 개요 및 종류**

&nbsp;&nbsp;&nbsp;&nbsp;선형회귀는 데이터 분석에서 가장 기초적인 예측 모델로, 독립 변수와 종속 변수 간의 선형 관계를 다룹니다. 즉, 특정 변수 x가 증가함에 따라 y가 증가하거나 감소하는 관계를 모델링합니다. 주로 연속형, 수치형 변수를 예측하는 데 사용됩니다. 가장 기본적인 형태는 다음과 같은 식으로 나타낼 수 있습니다:

#### &nbsp;&nbsp;&nbsp;&nbsp; **$$ y = wx + b $$**

&nbsp;&nbsp;&nbsp;&nbsp;여기서 w는 기울기(가중치), b는 절편(편향)입니다. 선형회귀의 목표는 예측 데이터와 실제 데이터 간의 차이를 최소화하는 최적의 w와 b를 찾아 예측력을 높이는 것입니다. 이러한 과정을 모델의 "최적화"라고 합니다.

&nbsp;&nbsp;&nbsp;&nbsp;선형회귀 모델은 일상생활에서도 널리 사용됩니다. 예를 들어, 키 x에 따른 몸무게 y, 공부 시간 x에 따른 시험 성적 y, 소득 x에 따른 여가비 지출 y 등 비교적 단순한 문제부터, 부동산 가격 y와 같이 여러 변수를 사용해야 하는 복잡한 문제까지 다양하게 적용할 수 있습니다.

&nbsp;&nbsp;&nbsp;&nbsp;선형회귀 모델의 종류는 다음과 같습니다:

#### &nbsp;&nbsp;&nbsp;&nbsp;**단순선형회귀**: 독립변수 x가 하나인 가장 기본적인 선형회귀 모델. $$ y = wx + b $$
#### &nbsp;&nbsp;&nbsp;&nbsp;**다중선형회귀**: 독립변수 x가 두 개 이상인 모델로, 가장 많이 사용되는 선형회귀 모델.  $$ y = w_1x_1 + w_2x_2 + \cdots + w_nx_n + b $$
#### &nbsp;&nbsp;&nbsp;&nbsp;**다항선형회귀**: 독립변수가 2차 함수 이상인 경우. $$ y = w_1x + w_2x^2 + \cdots + b $$

# **2. 비용함수**

&nbsp;&nbsp;&nbsp;&nbsp;선형회귀 모델을 최적화하기 위해 예측 데이터와 실제 데이터 간의 오차를 최소화하는 w와 b 값을 찾아야 합니다. 이를 위해 사용하는 것이 바로 비용함수(Cost Function)입니다. 비용함수는 실제 데이터와 예측 데이터 간의 오차를 측정하고, 이를 최소화하는 방향으로 모델을 최적화합니다.

&nbsp;&nbsp;&nbsp;&nbsp;선형회귀의 비용함수로 일반적으로 쓰이는 것들은 다음과 같습니다:

&nbsp;&nbsp;&nbsp;&nbsp;**MSE (Mean Squared Error)**: 평균 제곱 오차  
&nbsp;&nbsp;&nbsp;&nbsp;**RMSE (Root Mean Squared Error)**: 평균 제곱근 오차  
&nbsp;&nbsp;&nbsp;&nbsp;**MAE (Mean Absolute Error)**: 평균 절대 오차  
&nbsp;&nbsp;&nbsp;&nbsp;**R-Square (Coefficient of Determination)**: 결정 계수  

&nbsp;&nbsp;&nbsp;&nbsp;이러한 비용함수를 사용하여 최적의 w와 b 값을 찾고, 이를 통해 모델의 예측력을 높이는 것이 선형회귀 모델링의 일반적인 과정입니다.

# **3. 선형회귀의 가정**

&nbsp;&nbsp;&nbsp;&nbsp;선형회귀는 다음의 것들을 기본 전제로 하고 있습니다.

&nbsp;&nbsp;&nbsp;&nbsp;**1) 선형성**: 독립변수와 종속변수 사이에 선형적 관계가 어느 정도 있어야 제대로 된 모델링이 가능합니다. 
&nbsp;&nbsp;&nbsp;&nbsp;**2) 등분산성**: 오차가 특정 패턴을 보이지 않으며, 오차의 분산이 모든 수준의 독립 변수에 대해 일정해야 합니다.
&nbsp;&nbsp;&nbsp;&nbsp;**3) 정규성**: 오차항이 정규 분포를 따라야 합니다(외도 skewness가 있으면 안 됨).
&nbsp;&nbsp;&nbsp;&nbsp;**4) 독립성**: 독립변수 간 상관관계가 강하면 "다중공선성" 문제가 나타납니다.

```python
df[['sex', 'smoker', 'region']] = df[['sex', 'smoker', 'region']].astype('category')
df.dtypes
```

```python
##라벨 인코딩으로 수치형으로 변환
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
label.fit(df.sex.drop_duplicates())
df.sex = label.transform(df.sex)
label.fit(df.smoker.drop_duplicates())
df.smoker = label.transform(df.smoker)
label.fit(df.region.drop_duplicates())
df.region = label.transform(df.region)
df.dtypes
```

```python
f, ax = plt.subplots(1, 1, figsize=(10, 10))
ax = sns.heatmap(df.corr(), annot=True, cmap='cool')
#흡연, 그리고 그나마 나이 외에는 비용에 별다른 영향 미치는 변수가 없음
```

선형회귀 수행

```python
from sklearn.model_selection import train_test_split as holdout
from sklearn.linear_model import LinearRegression
from sklearn import metrics
x = df.drop(['charges'], axis = 1)
y = df['charges']
x_train, x_test, y_train, y_test = holdout(x, y, test_size=0.2, random_state =42)
Lin_reg = LinearRegression()
Lin_reg.fit(x_train, y_train)
y_train_pred = Lin_reg.predict(x_train)
y_test_pred = Lin_reg.predict(x_test)
print(Lin_reg.intercept_)
print(Lin_reg.coef_)
print(Lin_reg.score(x_test, y_test))
```

```python
##Evaluating the performance of the algorithm
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_test_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_test_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
```

### 다항식으로 변환해서 다항선형회귀 수행!
상관관계가 적은 'sex', 'region'를 독립변수에서 제외한 뒤 smoker 독립변수를 다항식으로 변환해서 선형회귀 수행

```python
from sklearn.preprocessing import PolynomialFeatures
x = df.drop(['charges', 'sex', 'region'], axis = 1)
y = df.charges
pol = PolynomialFeatures (degree = 2)
x_pol = pol.fit_transform(x)
x_train, x_test, y_train, y_test = holdout(x_pol, y, test_size=0.2, random_state=42)
Pol_reg = LinearRegression()
Pol_reg.fit(x_train, y_train)
y_train_pred = Pol_reg.predict(x_train)
y_test_pred = Pol_reg.predict(x_test)
print(Pol_reg.intercept_)
print(Pol_reg.coef_)
print(Pol_reg.score(x_test, y_test))
```

## 다항회귀로 했을 시 결과 더 좋음

```python
##Evaluating the performance of the algorithm
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_test_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_test_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
```

```python
##Predicting the charges
y_test_pred = Pol_reg.predict(x_test)
##Comparing the actual output values with the predicted values
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred})
df
```
