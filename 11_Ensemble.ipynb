## Ensenble Learning

### 1. 개요

&nbsp;&nbsp;&nbsp; 앙상블 기법은 여러 개의 개별 모델을 결합해 하나의 강력한 모델을 만드는 방법입니다. 이를 통해 단일 모델일 때보다 정확성을 높이고, 결과에 대한 일반화 가능성을 높일 수 있다는 장점이 있습니다. 그리고 최종 예측 결과는 수치형 변수의 경우 평균치로, 범주형의 경우 다수결로 결정해서 산출합니다.

### 2. 종류

#### 1. Bagging(*Boostrap Aggregating)

&nbsp;&nbsp;&nbsp; 원데이터 집합에서 크기가 같은 표본을 여러번 단순임의복원추출(bootstrap)해서 만든 모델들의 결과물을 집계(aggregating)하는 방법입니다. 각 모델은 모두 같은 것을 사용한다는 특징이 있으며 보통 의사결정나무를 사용합니다. 단순임의복원추출을 수행하기에 같은 데이터가 여러번 추출될 수도 있으며, 각 표본이 모집단의 특성을 잘 반영한다는 특성이 있습니다.

##### 2. Randonforest, Extra Tree

&nbsp;&nbsp;&nbsp; **1)랜덤포레스트 :** 의사결정나무에 기반하는 모델이자 배깅의 확장된 형태로 표본 추출뿐만 아니라 특성 선택에서도 랜덤과정이 추가된 모델입니다. 즉, 원데이터 집합에서 무작위로 데이터를 추출하는 것에 더해, 랜덤포레스트는 각 의사결정나무의 노드에서 분할기준으로 사용되는 피처도 무작위로 선택된 것만 사용한다는 특징이 있습니다. 단, 기존 의사결정나무와 같이 각 노드에서 최적의 분할을 찾기 위해 각 서브셋에서 최적의 분할 지점을 선택합니다. 랜덤과정이 추가되었기에 예측력이 강하지만 동시에 해석이 어렵다는 단점이 있습니다.

&nbsp;&nbsp;&nbsp; **2)엑스트라 트리 :** 배깅의 변형된 형태로, 데이터 샘플링 없이 각각의 모델이 전체 데이터를 사용한다는 차이점이 있습니다. 또한 노드의 분할기준으로 무작위로 선택한 피처만 사용한다는 점에서 랜덤포레스트와 비슷하나, 이 때 피처를 최적의 분할을 기준으로 선택하지 않고 무작위로 선택한다는 차이점이 있습니다.

##### 2. Voting : Hard Voting, Soft Voting, Averaging

&nbsp;&nbsp;&nbsp; 보팅은 기본적으로 같은 데이터로 여러 모델을 독립적으로 훈련시킨 뒤 각 모델의 예측을 결합해 최종 예측을 도출하는 모델입니다. 배깅이 원 데이터 집합에 대한 여러 샘플에 대해 같은 모델을 사용한다는 점에서 차이가 있습니다. 결과 도출은 분류 문제냐 회귀 문제냐에 따라 다른 방식을 사용하며, 분류 문제는 다시 두가지 방식으로 나눠서 결과를 도출합니다.

&nbsp;&nbsp;&nbsp; **1)하드보팅과 소프트보팅**

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; -분류문제에 사용합니다.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; -하드보팅은 각 모델의 결과에 대해 다수결 투표를 수행해서 최종 결과를 도출합니다.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; -소프트보팅은 각 모델의 결과를 평균내서 확률이 제일 높은 값을 최종 결과로 도출합니다.

&nbsp;&nbsp;&nbsp; **2)평균화 :** 회귀 문제에 사용되며, 각 모델의 예측값을 평균내서 최종 예측값을 도출합니다.

#### 3. Boosting

&nbsp;&nbsp;&nbsp; 부스팅은 배깅과 유사하지만, 재표본 과정에서 분류가 잘못된 데이터에 더 큰 가중치를 둔다는 차이점이 있습니다('잔차학습'). 즉, 배깅은 각 모델이 서로에게 영향을 미치지 않는 반면, 부스팅은 잘못 분류된 데이터에 집중해서 새로운 분류 규칙을 만드는 과정을 반복합니다. 이렇게 바로 앞선 모델이 예측할 때 남긴 오류를 바로 뒤의 모델이 학습해나가는 식으로 훈련 오차를 줄여나가면서 점차 강한 모델을 만들어가며, 보통 최종 결과를 도출할 때에도 마지막의 모델일 수록 비중을 더 크게 둡니다. 정확도는 높지만 이상치에 취약할 수 있다는 단점이 있습니다. 종류로는 AdaBoost(Adaptive Boosting), Gradient Boost, XGBoost, LightBoost 등이 있습니다. 이에 대해서는 추후 포스팅에서 다루도록 하겠습니다.

#### 4. Stacking

&nbsp;&nbsp;&nbsp; 스태킹은 여러 상이한 모델이 도출한 결과를 기반으로 최종 모델을 다시 학습시켜서 최종 메타모델을 생성하고 예측결과를 도출하는 모델입니다. 즉, 다른 모델들의 예측값들을 최종 메타 모델이 다시 학습해서 최종 예측치를 산출하는 모델입니다. 여러 모델의 장점을 활용할 수 있고 일반화 성능을 향상시킬 수 있다는 장점이 있습니다.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style("whitegrid")
plt.style.use("fivethirtyeight")
```

```python
df = pd.read_csv("/content/drive/MyDrive/코딩 공부 폴더/임시/diabetes_pima.csv")
df.head()
```

```python
df.info()
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)
df.describe()
```

```python
categorical_val = []
continous_val = []
for column in df.columns:
#     print('==============================')
#     print(f"{column} : {df[column].unique()}")
    if len(df[column].unique()) <= 10:
        categorical_val.append(column)
    else:
        continous_val.append(column)
```

```python
feature_columns = [
    'Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
    'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age'
]

for column in feature_columns:
    print("============================================")
    print(f"{column} ==> Missing zeros : {len(df.loc[df[column] == 0])}")
```

```python
from sklearn.impute import SimpleImputer


fill_values = SimpleImputer(missing_values=0, strategy="mean", copy=False)
df[feature_columns] = fill_values.fit_transform(df[feature_columns])

for column in feature_columns:
    print("============================================")
    print(f"{column} ==> Missing zeros : {len(df.loc[df[column] == 0])}")
```

```python
from sklearn.model_selection import train_test_split


X = df[feature_columns]
y = df.Outcome

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report


def evaluate(model, X_train, X_test, y_train, y_test):
    y_test_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)

    print("TRAINIG RESULTS: \n===============================")
    clf_report = pd.DataFrame(classification_report(y_train, y_train_pred, output_dict=True))
    print(f"CONFUSION MATRIX:\n{confusion_matrix(y_train, y_train_pred)}")
    print(f"ACCURACY SCORE:\n{accuracy_score(y_train, y_train_pred):.4f}")
    print(f"CLASSIFICATION REPORT:\n{clf_report}")

    print("TESTING RESULTS: \n===============================")
    clf_report = pd.DataFrame(classification_report(y_test, y_test_pred, output_dict=True))
    print(f"CONFUSION MATRIX:\n{confusion_matrix(y_test, y_test_pred)}")
    print(f"ACCURACY SCORE:\n{accuracy_score(y_test, y_test_pred):.4f}")
    print(f"CLASSIFICATION REPORT:\n{clf_report}")
```

# 배깅

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
bagging_clf = BaggingClassifier(tree, n_estimators=1500, random_state=42)
bagging_clf.fit(X_train, y_train)

evaluate(bagging_clf, X_train, X_test, y_train, y_test)
```

```python
scores = {
    'Bagging Classifier': {
        'Train': accuracy_score(y_train, bagging_clf.predict(X_train)),
        'Test': accuracy_score(y_test, bagging_clf.predict(X_test)),
    },
}
```

# 랜덤 포레스트

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(random_state=42, n_estimators=1000)
rf_clf.fit(X_train, y_train)
evaluate(rf_clf, X_train, X_test, y_train, y_test)
```

```python
scores['Random Forest'] = {
        'Train': accuracy_score(y_train, rf_clf.predict(X_train)),
        'Test': accuracy_score(y_test, rf_clf.predict(X_test)),
    }
```

# 엑스트라 트리

```python
from sklearn.ensemble import ExtraTreesClassifier

ex_tree_clf = ExtraTreesClassifier(n_estimators=1000, max_features=7, random_state=42)
ex_tree_clf.fit(X_train, y_train)
evaluate(ex_tree_clf, X_train, X_test, y_train, y_test)
```

```python
scores['Extra Tree'] = {
        'Train': accuracy_score(y_train, ex_tree_clf.predict(X_train)),
        'Test': accuracy_score(y_test, ex_tree_clf.predict(X_test)),
    }
```

# 부스팅

```python
from sklearn.ensemble import AdaBoostClassifier

ada_boost_clf = AdaBoostClassifier(n_estimators=30)
ada_boost_clf.fit(X_train, y_train)
evaluate(ada_boost_clf, X_train, X_test, y_train, y_test)
```

```python
scores['AdaBoost'] = {
        'Train': accuracy_score(y_train, ada_boost_clf.predict(X_train)),
        'Test': accuracy_score(y_test, ada_boost_clf.predict(X_test)),
    }
```

```python
from sklearn.ensemble import GradientBoostingClassifier

grad_boost_clf = GradientBoostingClassifier(n_estimators=100, random_state=42)
grad_boost_clf.fit(X_train, y_train)
evaluate(grad_boost_clf, X_train, X_test, y_train, y_test)
```

```python
scores['Gradient Boosting'] = {
        'Train': accuracy_score(y_train, grad_boost_clf.predict(X_train)),
        'Test': accuracy_score(y_test, grad_boost_clf.predict(X_test)),
    }
```

# 보팅

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

estimators = []
log_reg = LogisticRegression(solver='liblinear')
estimators.append(('Logistic', log_reg))

tree = DecisionTreeClassifier()
estimators.append(('Tree', tree))

svm_clf = SVC(gamma='scale')
estimators.append(('SVM', svm_clf))

voting = VotingClassifier(estimators=estimators)
voting.fit(X_train, y_train)

evaluate(voting, X_train, X_test, y_train, y_test)
```

```python
scores['Voting'] = {
        'Train': accuracy_score(y_train, voting.predict(X_train)),
        'Test': accuracy_score(y_test, voting.predict(X_test)),
    }
```

```python
scores_df = pd.DataFrame(scores)

scores_df.plot(kind='barh', figsize=(15, 8))
```
