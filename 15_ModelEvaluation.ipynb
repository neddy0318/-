# Ways to Evaluate Machine Learning Models

## 1. 개요

&nbsp;&nbsp; - 모델 성능 평가는 실제값과 모델에 의한 예측값을 비교하는 것으로, 과적합과 과소적합을 방지하고 최적의 모델을 찾기 위해 수행합니다.

## 2. 예측/회귀모델 평가지표

&nbsp;&nbsp; - 예측값과 실제값의 차이, 즉 오차를 구하는 것이기에 보통 값이 작을 수록 모델 성능이 좋다는 것을 의미합니다.

&nbsp;&nbsp; **1)MSE (Mean Squared Error)**

&nbsp;&nbsp;&nbsp; - 예측값과 실제값의 차이의 제곱의 평균입니다.

&nbsp;&nbsp;&nbsp; - 특징: 오차의 제곱을 사용하기 때문에 큰 오차에 더 큰 패널티를 부여합니다. 값이 작을수록 좋으며, 단위는 제곱된 값입니다.

$$ \text{MSE} = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2 $$

&nbsp;&nbsp;&nbsp;

&nbsp;&nbsp; **2)RMSE (Root Mean Squared Error)**

&nbsp;&nbsp;&nbsp; - MSE의 제곱근, 즉 오차의 제곱의 평균을 제곱근하여 원래의 단위로 되돌립니다.

&nbsp;&nbsp;&nbsp; - 특징: MSE보다 해석이 더 직관적이며, 오차의 단위를 원래의 단위와 동일하게 유지합니다. 값이 작을수록 좋습니다.

$$ \text{RMSE} = \sqrt{\text{MSE}} $$

&nbsp;&nbsp;&nbsp;

&nbsp;&nbsp; **3)MAE (Mean Absolute Error)**

&nbsp;&nbsp;&nbsp; - 예측값과 실제값의 차이의 절대값의 평균입니다.

&nbsp;&nbsp;&nbsp; - 특징: 오차의 절대값을 평균하여 계산하므로, 오차가 크거나 작은 것에 대해 균일하게 반응합니다. MSE보다 이상치에 덜 민감합니다. 값이 작을수록 좋습니다.

$$ \text{MAE} = \frac{1}{n} \sum_{i=1}^n |y_i - \hat{y}_i| $$

&nbsp;&nbsp;&nbsp;

&nbsp;&nbsp; **4)MAPE (Mean Absolute Percentage Error)**

&nbsp;&nbsp;&nbsp; - 예측값과 실제값의 차이의 절대값을 실제값으로 나눈 뒤 평균하여 백분율로 나타냅니다.

&nbsp;&nbsp;&nbsp; - 특징: 오차를 백분율로 나타내어 해석이 용이합니다. 실제값이 작을 때 오차 비율이 커질 수 있으므로, 이러한 상황에서는 MAPE가 비효율적일 수 있습니다. 값이 작을수록 좋습니다.


$$ \text{MAPE} = \frac{100\%}{n} \sum_{i=1}^n \left| \frac{y_i - \hat{y}_i}{y_i} \right| $$


## 3. 분류모델 평가지표

&nbsp;&nbsp; **1)정확도 (Accuracy)**

&nbsp;&nbsp;&nbsp; - 전체 예측 중에서 올바르게 예측한 비율을 의미합니다.

$$ \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN} $$

&nbsp;&nbsp; **2)오차행렬 (Confusion Matrix)**

&nbsp;&nbsp;&nbsp; - 분류 모델의 성능을 나타내는 표로, 각 클래스에 대해 True Positive, True Negative, False Positive, False Negative를 나타냅니다.

&nbsp;&nbsp;&nbsp; - 각 클래스의 예측 결과를 상세하게 나타내어 모델의 성능을 분석하는 데 유용합니다.


$$ 
\begin{matrix}
& \text{Predicted Positive} & \text{Predicted Negative} \\
\text{Actual Positive} & TP & FN \\
\text{Actual Negative} & FP & TN
\end{matrix}
$$


&nbsp;&nbsp; **3)정밀도 (Precision)**

&nbsp;&nbsp;&nbsp; - 모델이 Positive로 예측한 것 중 실제로 Positive인 비율을 의미합니다.

&nbsp;&nbsp;&nbsp; - 정밀도가 높으면 모델이 Positive로 예측한 것 중 실제 Positive인 경우가 많습니다. 값이 클수록 좋습니다.

$$ \text{Precision} = \frac{TP}{TP + FP} $$


&nbsp;&nbsp; **4)재현율 (Recall)**

&nbsp;&nbsp;&nbsp; - 실제 Positive 중에서 모델이 Positive로 올바르게 예측한 비율을 의미합니다.

&nbsp;&nbsp;&nbsp; - 재현율이 높으면 실제 Positive를 놓치지 않고 잘 찾아내는 모델입니다. 값이 클수록 좋습니다.

$$ \text{Recall} = \frac{TP}{TP + FN} $$


&nbsp;&nbsp; **5)F1-score**

&nbsp;&nbsp;&nbsp; - 정밀도와 재현율의 조화 평균. 두 지표의 균형을 평가합니다.

&nbsp;&nbsp;&nbsp; - 정밀도와 재현율이 모두 중요한 경우, 이 두 지표의 균형을 맞추는 데 유용합니다. 값이 클수록 좋습니다.

$$ \text{F1-score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}} $$



&nbsp;&nbsp; **6)ROC - AUC**

&nbsp;&nbsp;&nbsp; **(1)ROC 곡선**

![](https://velog.velcdn.com/images/zxxzx1515/post/a1114a22-b77a-4325-a02d-cb41ab079312/image.png)

&nbsp;&nbsp;&nbsp;&nbsp; - 정의: 다양한 임계값에서의 True Positive Rate (TPR)과 False Positive Rate (FPR)을 나타낸 곡선입니다. 예를 들어, 분류 임계값을 0.3, 0.5, 0.7 등으로 설정했을 때 각각의 TPR과 FPR 값을 계산해 그에 해당하는 점을 찍고 연결하는 식입니다.

  - **True Positive Rate (TPR)**:
    $$ \text{TPR} = \frac{TP}{TP + FN} $$
  - **False Positive Rate (FPR)**:
    $$ \text{FPR} = \frac{FP}{FP + TN} $$

&nbsp;&nbsp;&nbsp; **(2) AUC (Area Under the Curve)**

&nbsp;&nbsp;&nbsp;&nbsp; **정의**: ROC 곡선 아래 면적으로 모델의 분류 성능을 종합적으로 평가합니다.

&nbsp;&nbsp;&nbsp;&nbsp; **해석**:

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;  - **AUC = 0.5**: 모델이 무작위 추측과 같은 성능을 가진다는 의미입니다.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;  - **0.5 < AUC < 1**: 모델이 무작위 추측보다 더 나은 성능을 가진다는 뜻입니다.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;  - **AUC = 1**: 모델이 완벽한 분류 성능을 가지고 있다는 의미입니다.

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(data, test_size=0.1, random_state=42)    # Create, random train and test set


train_df_labels = train_df["Outcome"].copy()                                  # Drop labels
train_df= train_df.drop("Outcome", axis=1) 
```

```python
#Transformation Pipelines 

from sklearn.pipeline import Pipeline 
from sklearn.preprocessing import StandardScaler                       # This method makes normalization to our features

num_pipeline = Pipeline([('std_scaler', StandardScaler()), ])

train_prepared = num_pipeline.fit_transform(train_df)
```

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(train_prepared, train_df_labels)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(sgd_clf, train_prepared, train_df_labels, cv= 3, scoring='accuracy')
```

```python
prediction = sgd_clf.predict(train_prepared)
print("SGD Accuracy of Classifier: ", sgd_clf.score(train_prepared, train_df_labels))
```

```python
model = LogisticRegression()
model.fit(train_prepared, train_df_labels)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(model, train_prepared, train_df_labels, cv= 3, scoring='accuracy')
```

```python
prediction = model.predict(train_prepared)
print("LR Accuracy of Classifier: ", model.score(train_prepared, train_df_labels))
```

```python
from sklearn.svm import SVC 
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

poly_kernel_svm_clf = Pipeline([ ("scaler", StandardScaler()), 
                                ("svm_clf", SVC(kernel="poly", degree=3, coef0=1, C=5))
]) 

poly_kernel_svm_clf.fit(train_prepared, train_df_labels)
```

```python
print("SVM Accuracy of Classifier: ", poly_kernel_svm_clf.score(train_prepared, train_df_labels))
```

```python
from sklearn.model_selection import cross_val_predict 
y_train_pred = cross_val_predict(poly_kernel_svm_clf, train_prepared, train_df_labels, cv=3)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(train_df_labels, y_train_pred)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

print('Precision Score:',precision_score(train_df_labels, y_train_pred))
print('Recall Score:',recall_score(train_df_labels, y_train_pred))
print('F1 Score:',f1_score(train_df_labels, y_train_pred))
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(train_df_labels, y_train_pred)
```

```python
plt.plot(fpr, tpr, linewidth=2, label=None) 

plt.plot([0, 1], [0, 1], 'k--')
plt.text(0.12, 0.71, "Higher\nthreshold", color="#333333")
plt.xlabel('False Positive Rate (Fall-Out)')
plt.ylabel('True Positive Rate (Recall)')
plt.grid()
plt.axis([0, 1, 0, 1])
plt.legend(loc="lower right", fontsize=13)

plt.show()
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(train_df_labels, y_train_pred)
```
