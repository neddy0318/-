## 2. ÏòàÏ∏°/ÌöåÍ∑ÄÎ™®Îç∏ ÌèâÍ∞ÄÏßÄÌëú

&nbsp;&nbsp; - ÏòàÏ∏°Í∞íÍ≥º Ïã§Ï†úÍ∞íÏùò Ï∞®Ïù¥, Ï¶â Ïò§Ï∞®Î•º Íµ¨ÌïòÎäî Í≤ÉÏù¥Í∏∞Ïóê Î≥¥ÌÜµ Í∞íÏù¥ ÏûëÏùÑ ÏàòÎ°ù Î™®Îç∏ ÏÑ±Îä•Ïù¥ Ï¢ãÎã§Îäî Í≤ÉÏùÑ ÏùòÎØ∏Ìï©ÎãàÎã§.

&nbsp;&nbsp; **1)MSE (Mean Squared Error)**

&nbsp;&nbsp;&nbsp; - ÏòàÏ∏°Í∞íÍ≥º Ïã§Ï†úÍ∞íÏùò Ï∞®Ïù¥Ïùò Ï†úÍ≥±Ïùò ÌèâÍ∑†ÏûÖÎãàÎã§.

&nbsp;&nbsp;&nbsp; - ÌäπÏßï: Ïò§Ï∞®Ïùò Ï†úÍ≥±ÏùÑ ÏÇ¨Ïö©ÌïòÍ∏∞ ÎïåÎ¨∏Ïóê ÌÅ∞ Ïò§Ï∞®Ïóê Îçî ÌÅ∞ Ìå®ÎÑêÌã∞Î•º Î∂ÄÏó¨Ìï©ÎãàÎã§. Í∞íÏù¥ ÏûëÏùÑÏàòÎ°ù Ï¢ãÏúºÎ©∞, Îã®ÏúÑÎäî Ï†úÍ≥±Îêú Í∞íÏûÖÎãàÎã§.

$$ \text{MSE} = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2 $$

&nbsp;&nbsp;&nbsp;

&nbsp;&nbsp; **2)RMSE (Root Mean Squared Error)**

&nbsp;&nbsp;&nbsp; - MSEÏùò Ï†úÍ≥±Í∑º, Ï¶â Ïò§Ï∞®Ïùò Ï†úÍ≥±Ïùò ÌèâÍ∑†ÏùÑ Ï†úÍ≥±Í∑ºÌïòÏó¨ ÏõêÎûòÏùò Îã®ÏúÑÎ°ú ÎêòÎèåÎ¶ΩÎãàÎã§.

&nbsp;&nbsp;&nbsp; - ÌäπÏßï: MSEÎ≥¥Îã§ Ìï¥ÏÑùÏù¥ Îçî ÏßÅÍ¥ÄÏ†ÅÏù¥Î©∞, Ïò§Ï∞®Ïùò Îã®ÏúÑÎ•º ÏõêÎûòÏùò Îã®ÏúÑÏôÄ ÎèôÏùºÌïòÍ≤å Ïú†ÏßÄÌï©ÎãàÎã§. Í∞íÏù¥ ÏûëÏùÑÏàòÎ°ù Ï¢ãÏäµÎãàÎã§.

$$ \text{RMSE} = \sqrt{\text{MSE}} $$

&nbsp;&nbsp;&nbsp;

&nbsp;&nbsp; **3)MAE (Mean Absolute Error)**

&nbsp;&nbsp;&nbsp; - ÏòàÏ∏°Í∞íÍ≥º Ïã§Ï†úÍ∞íÏùò Ï∞®Ïù¥Ïùò Ï†àÎåÄÍ∞íÏùò ÌèâÍ∑†ÏûÖÎãàÎã§.

&nbsp;&nbsp;&nbsp; - ÌäπÏßï: Ïò§Ï∞®Ïùò Ï†àÎåÄÍ∞íÏùÑ ÌèâÍ∑†ÌïòÏó¨ Í≥ÑÏÇ∞ÌïòÎØÄÎ°ú, Ïò§Ï∞®Í∞Ä ÌÅ¨Í±∞ÎÇò ÏûëÏùÄ Í≤ÉÏóê ÎåÄÌï¥ Í∑†ÏùºÌïòÍ≤å Î∞òÏùëÌï©ÎãàÎã§. MSEÎ≥¥Îã§ Ïù¥ÏÉÅÏπòÏóê Îçú ÎØºÍ∞êÌï©ÎãàÎã§. Í∞íÏù¥ ÏûëÏùÑÏàòÎ°ù Ï¢ãÏäµÎãàÎã§.

$$ \text{MAE} = \frac{1}{n} \sum_{i=1}^n |y_i - \hat{y}_i| $$

&nbsp;&nbsp;&nbsp;

&nbsp;&nbsp; **4)MAPE (Mean Absolute Percentage Error)**

&nbsp;&nbsp;&nbsp; - ÏòàÏ∏°Í∞íÍ≥º Ïã§Ï†úÍ∞íÏùò Ï∞®Ïù¥Ïùò Ï†àÎåÄÍ∞íÏùÑ Ïã§Ï†úÍ∞íÏúºÎ°ú ÎÇòÎàà Îí§ ÌèâÍ∑†ÌïòÏó¨ Î∞±Î∂ÑÏú®Î°ú ÎÇòÌÉÄÎÉÖÎãàÎã§.

&nbsp;&nbsp;&nbsp; - ÌäπÏßï: Ïò§Ï∞®Î•º Î∞±Î∂ÑÏú®Î°ú ÎÇòÌÉÄÎÇ¥Ïñ¥ Ìï¥ÏÑùÏù¥ Ïö©Ïù¥Ìï©ÎãàÎã§. Ïã§Ï†úÍ∞íÏù¥ ÏûëÏùÑ Îïå Ïò§Ï∞® ÎπÑÏú®Ïù¥ Ïª§Ïßà Ïàò ÏûàÏúºÎØÄÎ°ú, Ïù¥Îü¨Ìïú ÏÉÅÌô©ÏóêÏÑúÎäî MAPEÍ∞Ä ÎπÑÌö®Ïú®Ï†ÅÏùº Ïàò ÏûàÏäµÎãàÎã§. Í∞íÏù¥ ÏûëÏùÑÏàòÎ°ù Ï¢ãÏäµÎãàÎã§.


$$ \text{MAPE} = \frac{100\%}{n} \sum_{i=1}^n \left| \frac{y_i - \hat{y}_i}{y_i} \right| $$


## 3. Î∂ÑÎ•òÎ™®Îç∏ ÌèâÍ∞ÄÏßÄÌëú

&nbsp;&nbsp; **1)Ï†ïÌôïÎèÑ (Accuracy)**

&nbsp;&nbsp;&nbsp; - Ï†ÑÏ≤¥ ÏòàÏ∏° Ï§ëÏóêÏÑú Ïò¨Î∞îÎ•¥Í≤å ÏòàÏ∏°Ìïú ÎπÑÏú®ÏùÑ ÏùòÎØ∏Ìï©ÎãàÎã§.

$$ \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN} $$

&nbsp;&nbsp; **2)Ïò§Ï∞®ÌñâÎ†¨ (Confusion Matrix)**

&nbsp;&nbsp;&nbsp; - Î∂ÑÎ•ò Î™®Îç∏Ïùò ÏÑ±Îä•ÏùÑ ÎÇòÌÉÄÎÇ¥Îäî ÌëúÎ°ú, Í∞Å ÌÅ¥ÎûòÏä§Ïóê ÎåÄÌï¥ True Positive, True Negative, False Positive, False NegativeÎ•º ÎÇòÌÉÄÎÉÖÎãàÎã§.

&nbsp;&nbsp;&nbsp; - Í∞Å ÌÅ¥ÎûòÏä§Ïùò ÏòàÏ∏° Í≤∞Í≥ºÎ•º ÏÉÅÏÑ∏ÌïòÍ≤å ÎÇòÌÉÄÎÇ¥Ïñ¥ Î™®Îç∏Ïùò ÏÑ±Îä•ÏùÑ Î∂ÑÏÑùÌïòÎäî Îç∞ Ïú†Ïö©Ìï©ÎãàÎã§.


$$ 
\begin{matrix}
& \text{Predicted Positive} & \text{Predicted Negative} \\
\text{Actual Positive} & TP & FN \\
\text{Actual Negative} & FP & TN
\end{matrix}
$$


&nbsp;&nbsp; **3)Ï†ïÎ∞ÄÎèÑ (Precision)**

&nbsp;&nbsp;&nbsp; - Î™®Îç∏Ïù¥ PositiveÎ°ú ÏòàÏ∏°Ìïú Í≤É Ï§ë Ïã§Ï†úÎ°ú PositiveÏù∏ ÎπÑÏú®ÏùÑ ÏùòÎØ∏Ìï©ÎãàÎã§.

&nbsp;&nbsp;&nbsp; - Ï†ïÎ∞ÄÎèÑÍ∞Ä ÎÜíÏúºÎ©¥ Î™®Îç∏Ïù¥ PositiveÎ°ú ÏòàÏ∏°Ìïú Í≤É Ï§ë Ïã§Ï†ú PositiveÏù∏ Í≤ΩÏö∞Í∞Ä ÎßéÏäµÎãàÎã§. Í∞íÏù¥ ÌÅ¥ÏàòÎ°ù Ï¢ãÏäµÎãàÎã§.

$$ \text{Precision} = \frac{TP}{TP + FP} $$


&nbsp;&nbsp; **4)Ïû¨ÌòÑÏú® (Recall)**

&nbsp;&nbsp;&nbsp; - Ïã§Ï†ú Positive Ï§ëÏóêÏÑú Î™®Îç∏Ïù¥ PositiveÎ°ú Ïò¨Î∞îÎ•¥Í≤å ÏòàÏ∏°Ìïú ÎπÑÏú®ÏùÑ ÏùòÎØ∏Ìï©ÎãàÎã§.

&nbsp;&nbsp;&nbsp; - Ïû¨ÌòÑÏú®Ïù¥ ÎÜíÏúºÎ©¥ Ïã§Ï†ú PositiveÎ•º ÎÜìÏπòÏßÄ ÏïäÍ≥† Ïûò Ï∞æÏïÑÎÇ¥Îäî Î™®Îç∏ÏûÖÎãàÎã§. Í∞íÏù¥ ÌÅ¥ÏàòÎ°ù Ï¢ãÏäµÎãàÎã§.

$$ \text{Recall} = \frac{TP}{TP + FN} $$


&nbsp;&nbsp; **5)F1-score**

&nbsp;&nbsp;&nbsp; - Ï†ïÎ∞ÄÎèÑÏôÄ Ïû¨ÌòÑÏú®Ïùò Ï°∞Ìôî ÌèâÍ∑†. Îëê ÏßÄÌëúÏùò Í∑†ÌòïÏùÑ ÌèâÍ∞ÄÌï©ÎãàÎã§.

&nbsp;&nbsp;&nbsp; - Ï†ïÎ∞ÄÎèÑÏôÄ Ïû¨ÌòÑÏú®Ïù¥ Î™®Îëê Ï§ëÏöîÌïú Í≤ΩÏö∞, Ïù¥ Îëê ÏßÄÌëúÏùò Í∑†ÌòïÏùÑ ÎßûÏ∂îÎäî Îç∞ Ïú†Ïö©Ìï©ÎãàÎã§. Í∞íÏù¥ ÌÅ¥ÏàòÎ°ù Ï¢ãÏäµÎãàÎã§.

$$ \text{F1-score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}} $$



&nbsp;&nbsp; **6)ROC - AUC**

&nbsp;&nbsp;&nbsp; **(1)ROC Í≥°ÏÑ†**

ÏòàÎ•º Îì§Ïñ¥, Î∂ÑÎ•ò ÏûÑÍ≥ÑÍ∞íÏùÑ 0.3, 0.5, 0.7 Îì±ÏúºÎ°ú ÏÑ§Ï†ïÌñàÏùÑ Îïå Í∞ÅÍ∞ÅÏùò TPRÍ≥º FPR Í∞íÏùÑ Í≥ÑÏÇ∞Ìï¥ Í∑∏Ïóê Ìï¥ÎãπÌïòÎäî Ï†êÏùÑ Ï∞çÍ≥† Ïó∞Í≤∞ÌïòÎäî ÏãùÏûÖÎãàÎã§.

  - **True Positive Rate (TPR)**:
    $$ \text{TPR} = \frac{TP}{TP + FN} $$

    $$ \text{FPR} = \frac{FP}{FP + TN} $$

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(data, test_size=0.1, random_state=42)    # Create, random train and test set


train_df_labels = train_df["Outcome"].copy()                                  # Drop labels
train_df= train_df.drop("Outcome", axis=1) 
```

```python
#Transformation Pipelines 

from sklearn.pipeline import Pipeline 
from sklearn.preprocessing import StandardScaler                       # This method makes normalization to our features

num_pipeline = Pipeline([('std_scaler', StandardScaler()), ])

train_prepared = num_pipeline.fit_transform(train_df)
```

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(train_prepared, train_df_labels)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(sgd_clf, train_prepared, train_df_labels, cv= 3, scoring='accuracy')
```

```python
prediction = sgd_clf.predict(train_prepared)
print("SGD Accuracy of Classifier: ", sgd_clf.score(train_prepared, train_df_labels))
```

```python
model = LogisticRegression()
model.fit(train_prepared, train_df_labels)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(model, train_prepared, train_df_labels, cv= 3, scoring='accuracy')
```

```python
prediction = model.predict(train_prepared)
print("LR Accuracy of Classifier: ", model.score(train_prepared, train_df_labels))
```

```python
from sklearn.svm import SVC 
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

poly_kernel_svm_clf = Pipeline([ ("scaler", StandardScaler()), 
                                ("svm_clf", SVC(kernel="poly", degree=3, coef0=1, C=5))
]) 

poly_kernel_svm_clf.fit(train_prepared, train_df_labels)
```

```python
print("SVM Accuracy of Classifier: ", poly_kernel_svm_clf.score(train_prepared, train_df_labels))
```

```python
from sklearn.model_selection import cross_val_predict 
y_train_pred = cross_val_predict(poly_kernel_svm_clf, train_prepared, train_df_labels, cv=3)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(train_df_labels, y_train_pred)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

print('Precision Score:',precision_score(train_df_labels, y_train_pred))
print('Recall Score:',recall_score(train_df_labels, y_train_pred))
print('F1 Score:',f1_score(train_df_labels, y_train_pred))
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(train_df_labels, y_train_pred)
```

```python
plt.plot(fpr, tpr, linewidth=2, label=None) 

plt.plot([0, 1], [0, 1], 'k--')
plt.text(0.12, 0.71, "Higher\nthreshold", color="#333333")
plt.xlabel('False Positive Rate (Fall-Out)')
plt.ylabel('True Positive Rate (Recall)')
plt.grid()
plt.axis([0, 1, 0, 1])
plt.legend(loc="lower right", fontsize=13)

plt.show()
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(train_df_labels, y_train_pred)
```
