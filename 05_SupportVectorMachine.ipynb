### 4. Ïª§ÎÑê

&nbsp;&nbsp;&nbsp; ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã†ÏúºÎ°ú ÎπÑÏÑ†Ìòï Îç∞Ïù¥ÌÑ∞Î•º Î∂ÑÎ•òÌï† Îïå, Ïö∞ÏÑ† ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞Î•º Îã§Ï∞®Ïõê Í≥µÍ∞ÑÏóê Îß§ÌïëÌï¥Ïïº Î∂ÑÎ•òÍ∞Ä Í∞ÄÎä•Ìï©ÎãàÎã§. ## 5. Í≥µÏãù

### ÎßàÏßÑÏùò Í∏∏Ïù¥ Íµ¨ÌïòÎäî Í≥µÏãù

&nbsp;&nbsp;&nbsp;ÎßàÏßÑÏùò Í∏∏Ïù¥Îäî Í≤∞Ï†ï Í≤ΩÍ≥ÑÏôÄ ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ ÏÇ¨Ïù¥Ïùò Í±∞Î¶¨Î°ú Ï†ïÏùòÎê©ÎãàÎã§. Ïù¥ Í±∞Î¶¨Îäî Îã§ÏùåÍ≥º Í∞ôÏùÄ Í≥µÏãùÏúºÎ°ú Í≥ÑÏÇ∞Îê©ÎãàÎã§:

$$
\text{ÎßàÏßÑ} = \frac{2}{\|\mathbf{w}\|}
$$

&nbsp;&nbsp;&nbsp;Ïó¨Í∏∞ÏÑú \(\mathbf{w}\)Îäî Í≤∞Ï†ï Í≤ΩÍ≥ÑÏùò Î≤ïÏÑ† Î≤°ÌÑ∞(normal vector)ÏûÖÎãàÎã§. \(\|\mathbf{w}\|\)Îäî Î≤°ÌÑ∞ \(\mathbf{w}\)Ïùò Ïú†ÌÅ¥Î¶¨Îìú ÎÖ∏Î¶Ñ(Euclidean norm) ÎòêÎäî ÌÅ¨Í∏∞Î•º ÏùòÎØ∏Ìï©ÎãàÎã§.

### Ïª§ÎÑê Ìï®Ïàò Ïú†ÎèÑ Í≥µÏãù

&nbsp;&nbsp;&nbsp;ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã†ÏóêÏÑúÎäî ÎπÑÏÑ†Ìòï Îç∞Ïù¥ÌÑ∞Î•º Î∂ÑÎ•òÌïòÍ∏∞ ÏúÑÌï¥ Ïª§ÎÑê Ìï®ÏàòÎ•º ÏÇ¨Ïö©ÌïòÏó¨ ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞Î•º Í≥†Ï∞®Ïõê Í≥µÍ∞ÑÏúºÎ°ú Î≥ÄÌôòÌï©ÎãàÎã§. Ïª§ÎÑê Ìï®Ïàò \( K(\mathbf{x}_i, \mathbf{x}_j) \)Îäî Îëê Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏ \(\mathbf{x}_i\)ÏôÄ \(\mathbf{x}_j\) ÏÇ¨Ïù¥Ïùò Ïú†ÏÇ¨ÎèÑÎ•º Í≥ÑÏÇ∞Ìï©ÎãàÎã§. Ïª§ÎÑê Ìä∏Î¶≠ÏùÑ ÏÇ¨Ïö©ÌïòÎ©¥ Í≥†Ï∞®Ïõê Í≥µÍ∞ÑÏóêÏÑúÏùò ÎÇ¥Ï†ÅÏùÑ ÏßÅÏ†ë Í≥ÑÏÇ∞ÌïòÏßÄ ÏïäÍ≥†ÎèÑ Î∂ÑÎ•ò Î¨∏Ï†úÎ•º Ìï¥Í≤∞Ìï† Ïàò ÏûàÏäµÎãàÎã§.

&nbsp;&nbsp;&nbsp;ÎåÄÌëúÏ†ÅÏù∏ Ïª§ÎÑê Ìï®ÏàòÎäî Îã§ÏùåÍ≥º Í∞ôÏäµÎãàÎã§:

&nbsp;&nbsp;&nbsp; **1)ÏÑ†Ìòï Ïª§ÎÑê**:
$$
K(\mathbf{x}_i, \mathbf{x}_j) = \mathbf{x}_i \cdot \mathbf{x}_j
$$

&nbsp;&nbsp;&nbsp; **2)Îã§Ìï≠Ïãù Ïª§ÎÑê**:
$$
K(\mathbf{x}_i, \mathbf{x}_j) = (\mathbf{x}_i \cdot \mathbf{x}_j + c)^d
$$
Ïó¨Í∏∞ÏÑú \(c\)Îäî ÏÉÅÏàò, \(d\)Îäî Îã§Ìï≠ÏãùÏùò Ï∞®ÏàòÏûÖÎãàÎã§.

&nbsp;&nbsp;&nbsp; **3)Í∞ÄÏö∞ÏãúÏïà RBF Ïª§ÎÑê**:
$$
K(\mathbf{x}_i, \mathbf{x}_j) = \exp\left(-\frac{\|\mathbf{x}_i - \mathbf{x}_j\|^2}{2\sigma^2}\right)
$$
Ïó¨Í∏∞ÏÑú \(\sigma\)Îäî Í∞ÄÏö∞ÏãúÏïà Ìï®ÏàòÏùò ÌëúÏ§Ä Ìé∏Ï∞®ÏûÖÎãàÎã§.

&nbsp;&nbsp;&nbsp; **4)ÏãúÍ∑∏Î™®Ïù¥Îìú Ïª§ÎÑê**:
$$
K(\mathbf{x}_i, \mathbf{x}_j) = \tanh(\kappa \mathbf{x}_i \cdot \mathbf{x}_j + \theta)
$$

***KDE*** 

Kernel density estimation (KDE)ÏùÄ ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò(Probability Density Function, PDF)Î•º Ï∂îÏ†ïÌïòÍ∏∞ ÏúÑÌïú ÎπÑÎ™®ÏàòÏ†Å Î∞©Î≤ïÏûÖÎãàÎã§. Îç∞Ïù¥ÌÑ∞Í∞Ä Ïñ¥Îñ§ Î≥ÄÏàòÏóêÏÑú Ïñ¥ÎñªÍ≤å Î∂ÑÌè¨ÎêòÏñ¥ ÏûàÎäîÏßÄÎ•º ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú ÌëúÌòÑÌïòÎäî Îç∞ ÏÇ¨Ïö©Îê©ÎãàÎã§.

Î™®Îì† Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Ïóê ÎåÄÌïú Ïª§ÎÑê Ìï®Ïàò Ï†ÅÏö©: Í∞Å Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Ïóê ÎåÄÌï¥ Ïª§ÎÑê Ìï®ÏàòÎ•º Ï†ÅÏö©ÌïòÍ≥†, Ïù¥Îì§ÏùÑ Î™®Îëê Ìï©ÌïòÏó¨ Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ÏÖãÏùò ÌôïÎ•† Î∞ÄÎèÑ Ìï®ÏàòÎ•º Ï∂îÏ†ïÌï©ÎãàÎã§.

Ïª§ÎÑê Ìï®ÏàòÏùò ÎåÄÏó≠Ìè≠ ÏÑ§Ï†ï: KDEÏùò Í≤∞Í≥ºÎäî ÎåÄÏó≠Ìè≠ ÏÑ†ÌÉùÏóê Îî∞Îùº Îã¨ÎùºÏßëÎãàÎã§. ÎåÄÏó≠Ìè≠Ïù¥ ÏûëÏúºÎ©¥ Ï∂îÏ†ïÎêú ÌôïÎ•† Î∞ÄÎèÑ Ìï®ÏàòÎäî Îæ∞Ï°±ÌïòÍ≥† Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏ Ï£ºÏúÑÏóêÏÑú Í∏âÍ≤©ÌïòÍ≤å Î≥ÄÌôîÌï©ÎãàÎã§. ÎåÄÏó≠Ìè≠Ïù¥ ÌÅ¨Î©¥ Î≥¥Îã§ Î∂ÄÎìúÎüΩÍ≥† ÏôÑÎßåÌïú ÌôïÎ•† Î∞ÄÎèÑ Ï∂îÏ†ïÏù¥ Ïù¥Î£®Ïñ¥ÏßëÎãàÎã§.

Îî∞ÎùºÏÑú Kernel Density Estimation (KDE) plotÏùÄ Ï£ºÏñ¥ÏßÑ Îç∞Ïù¥ÌÑ∞Ïùò Î≥ÄÏàò(Ïó¨Í∏∞ÏÑúÎäî 'time'Í≥º 'age' ÌäπÏÑ±)Ïùò Î∂ÑÌè¨Î•º Î∂ÄÎìúÎüΩÍ≤å ÎÇòÌÉÄÎÇ¥Îäî Í∑∏ÎûòÌîÑÏûÖÎãàÎã§. Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Ïùò Î∞ÄÎèÑÍ∞Ä ÎÜíÏùÄ ÏßÄÏ†êÏùÄ KDE plot ÏÉÅÏóêÏÑú Îçî ÎÜíÏùÄ Í∞íÏúºÎ°ú ÎÇòÌÉÄÎÇòÎ©∞, Î≥ÄÏàòÍ∞Ä Ïñ¥ÎñªÍ≤å Î∂ÑÌè¨ÎêòÏñ¥ ÏûàÎäîÏßÄ ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú Ïûò Ïù¥Ìï¥Ìï† Ïàò ÏûàÏäµÎãàÎã§.

```python
X=data_df.drop(["DEATH_EVENT"],axis=1)
y=data_df["DEATH_EVENT"]
```

```python
col_names = list(X.columns)
s_scaler = preprocessing.StandardScaler()
X_scaled = s_scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns = col_names)
X_scaled.describe().T
```

```python
#Plotting the scaled features using boxen plots
colors =["#CD5C5C","#F08080","#FA8072","#E9967A","#FFA07A"]
plt.figure(figsize=(20,10))
sns.boxenplot(data = X_scaled,palette = colors)
plt.xticks(rotation=60)
plt.show()
```

***Boxen Plot***

SeabornÏùò boxenplotÏùÄ Í∏∞Î≥∏Ï†ÅÏúºÎ°ú boxplotÍ≥º ÎπÑÏä∑ÌïòÏßÄÎßå, Ï£ºÎ°ú Îç∞Ïù¥ÌÑ∞Ïùò Î∂ÑÌè¨Î•º Îçî ÏÑ∏Î∂ÄÏ†ÅÏúºÎ°ú ÏãúÍ∞ÅÌôîÌïòÎäî Îç∞ ÏÇ¨Ïö©Îê©ÎãàÎã§. ÏùºÎ∞òÏ†ÅÏù∏ boxplotÏùÄ Ï£ºÎ°ú Ï§ëÏïôÍ∞íÍ≥º ÏÇ¨Î∂ÑÏúÑÏàò(1ÏÇ¨Î∂ÑÏúÑÏàòÏôÄ 3ÏÇ¨Î∂ÑÏúÑÏàò)Î•º Î≥¥Ïó¨Ï£ºÎ©∞, Ïù¥ÏÉÅÏπòÎ•º ÌëúÌòÑÌï† Ïàò ÏûàÏäµÎãàÎã§. Î∞òÎ©¥Ïóê boxenplotÏùÄ 'letter-value' ÌîåÎ°ØÏù¥ÎùºÍ≥†ÎèÑ Î∂àÎ¶¨Î©∞, Îçî ÎßéÏùÄ ÏÇ¨Î∂ÑÏúÑÏàòÎ•º Ìè¨Ìï®ÌïòÏó¨ Îç∞Ïù¥ÌÑ∞Ïùò Î∂ÑÌè¨Î•º ÏÉÅÏÑ∏ÌïòÍ≤å Î≥¥Ïó¨Ï§çÎãàÎã§.

Îî∞ÎùºÏÑú boxenplotÏùÄ Ïó¨Îü¨ Í∞úÏùò Î∞ïÏä§Î•º ÏÉùÏÑ±ÌïòÏó¨ Í∞Å Îç∞Ïù¥ÌÑ∞ Î∂ÑÌè¨Ïùò ÏÑ∏Î∂ÄÏ†ÅÏù∏ Ï†ïÎ≥¥Î•º Ï†úÍ≥µÌïòÎäî Í≤ÉÏù¥ ÏïÑÎãàÎùº, Îã®Ïùº Î∞ïÏä§ ÎÇ¥ÏóêÏÑú Î≥¥Îã§ ÎßéÏùÄ ÏÇ¨Î∂ÑÏúÑÏàò Ï†ïÎ≥¥Î•º Ï†úÍ≥µÌïòÎäî Î∞©ÏãùÏúºÎ°ú ÏÑ§Í≥ÑÎêòÏóàÏäµÎãàÎã§.

```python
#spliting variables into training and test sets
X_train, X_test, y_train,y_test = train_test_split(X_scaled,y,test_size=0.30,random_state=25)
```

### Î™®Îç∏ Íµ¨Ï∂ï

```python
model1 = svm.SVC()
model1.fit(X_train, y_train)
y_pred = model1.predict(X_test)
model1.score(X_test, y_test)
```

```python
from sklearn.metrics import precision_score, recall_score, confusion_matrix, classification_report, accuracy_score, f1_score
```

```python
print(classification_report(y_test, y_pred))
```

***classification_report***
