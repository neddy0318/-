## 서포트 벡터 머신(SVM, Support Vector Machine)

### 1. 서포트 벡터 머신이란?

&nbsp;&nbsp;&nbsp; **서포트 벡터 머신(SVM)**이란 데이터들을 이진분류하는 분류선을 기반으로 회귀와 분류 모두를 수행하는 모델입니다. 좀 더 전문적으로 말하자면, 분류선을 통해 마진을 최대화하고, 이를 기반으로 분류 혹은 회귀 성능의 일반화 가능성을 극대화하는 것을 목표로 합니다. 여기서 말하는 **마진**이 무엇인지, 그리고 모델링 과정이 구체적으로 어떻게 이루어지는지 아래에서 살펴보겠습니다.

### 2. 주요 용어

&nbsp;&nbsp;&nbsp; **1) 결정 경계(=초평면, hyperplane):** 데이터를 이진분류하는 기준이 되는 선으로, 데이터가 \( n \)차원에 있으면 초평면은 \( n-1 \) 차원의 하위공간(sub-space)에 존재하게 됩니다. 가령 2차원에서는 선(line), 3차원에서는 평면(plane)이, 그보다 더 높은 차원에서는 '초평면'이 됩니다. 서포트 벡터 머신은 이를 기반으로 분류 혹은 회귀를 수행합니다.

&nbsp;&nbsp;&nbsp; **2) 서포트 벡터:** 일반적으로 벡터는 데이터 포인트를 의미합니다. 가령 2차원에서는 각 데이터 포인트가 \((x_1, y_1)\)으로 표현되는데요. 서포트 벡터는 결정 경계와 가장 근접해있으면서 이를 정의하거나 지지(support)하는 데이터 포인트들입니다.

&nbsp;&nbsp;&nbsp; **3) 마진:** 서포트 벡터와 결정 경계 사이의 거리를 의미합니다.

### 3. 마진 종류

&nbsp;&nbsp;&nbsp; 그렇다면 적절한 결정 경계를 결정할 **마진**은 어떻게 설정하는 걸까요? 이는 이상치를 얼마나 허용할지에 달려 있으며, 그에 따라 마진을 **소프트 마진**과 **하드 마진**으로 구분합니다. 이상치를 얼마나 허용할지는 파라미터 \( C \) (cost)로 조절하며, \( C \) 값이 클수록 이상치를 적게, 작을수록 많이 허용합니다.

&nbsp;&nbsp;&nbsp; **1) 하드 마진:** 하드 마진은 모든 데이터가 결정 경계의 마진 밖으로 위치하도록 하는 가중치(기울기)를 찾는 방법입니다. 그에 따라 서포트 벡터와 결정 경계 사이가 좁으며, 이상치가 없이 선형 분리가 가능한 데이터에 대해 오분류가 없을 수 있다는 장점이 있습니다. 하지만 그렇지 않은 데이터에 대해서는 적합하지 않으며, 과적합 문제가 발생할 수 있습니다.

&nbsp;&nbsp;&nbsp; **2) 소프트 마진:** 소프트 마진은 결정 경계 마진 안에 위치하는 데이터 수를 제한하면서 마진 폭이 최대가 되도록 하는 가중치(기울기)를 찾는 방법입니다. 이상치에 강하며, 비선형적으로 분리 가능한 데이터에 더욱 적합합니다. 하지만 과소적합 문제가 발생할 가능성이 있습니다.


### 4. 커널

&nbsp;&nbsp;&nbsp; 서포트 벡터 머신으로 비선형 데이터를 분류할 때, 우선 입력 데이터를 다차원 공간에 매핑해야 분류가 가능합니다. 이를 위해 커널 함수를 활용해 계산량을 줄입니다. 커널 기법에는 다음과 같은 종류가 있습니다.

&nbsp;&nbsp;&nbsp; **1) 선형 커널:** 1차원의 기본 커널로, 텍스트 분류에 주로 사용합니다.

&nbsp;&nbsp;&nbsp; **2) 다항식 커널:** 다항 특성을 활용한 함수로, 지정한 차수까지 변환해서 확장할 수 있습니다.

&nbsp;&nbsp;&nbsp; **3) 가우시안 RBF 커널:** 유사도 특성을 추가하는 효과를 활용한 커널로, 데이터에 대한 사전 지식이 없는 경우 주로 사용합니다.



## 5. 공식

### 마진의 길이 구하는 공식

&nbsp;&nbsp;&nbsp;마진의 길이는 결정 경계와 서포트 벡터 사이의 거리로 정의됩니다. 이 거리는 다음과 같은 공식으로 계산됩니다:

$$
\text{마진} = \frac{2}{\|\mathbf{w}\|}
$$

&nbsp;&nbsp;&nbsp;여기서 \(\mathbf{w}\)는 결정 경계의 법선 벡터(normal vector)입니다. \(\|\mathbf{w}\|\)는 벡터 \(\mathbf{w}\)의 유클리드 노름(Euclidean norm) 또는 크기를 의미합니다.

### 커널 함수 유도 공식

&nbsp;&nbsp;&nbsp;서포트 벡터 머신에서는 비선형 데이터를 분류하기 위해 커널 함수를 사용하여 입력 데이터를 고차원 공간으로 변환합니다. 커널 함수 \( K(\mathbf{x}_i, \mathbf{x}_j) \)는 두 데이터 포인트 \(\mathbf{x}_i\)와 \(\mathbf{x}_j\) 사이의 유사도를 계산합니다. 커널 트릭을 사용하면 고차원 공간에서의 내적을 직접 계산하지 않고도 분류 문제를 해결할 수 있습니다.

&nbsp;&nbsp;&nbsp;대표적인 커널 함수는 다음과 같습니다:

&nbsp;&nbsp;&nbsp; **1)선형 커널**:
$$
K(\mathbf{x}_i, \mathbf{x}_j) = \mathbf{x}_i \cdot \mathbf{x}_j
$$

&nbsp;&nbsp;&nbsp; **2)다항식 커널**:
$$
K(\mathbf{x}_i, \mathbf{x}_j) = (\mathbf{x}_i \cdot \mathbf{x}_j + c)^d
$$
여기서 \(c\)는 상수, \(d\)는 다항식의 차수입니다.

&nbsp;&nbsp;&nbsp; **3)가우시안 RBF 커널**:
$$
K(\mathbf{x}_i, \mathbf{x}_j) = \exp\left(-\frac{\|\mathbf{x}_i - \mathbf{x}_j\|^2}{2\sigma^2}\right)
$$
여기서 \(\sigma\)는 가우시안 함수의 표준 편차입니다.

&nbsp;&nbsp;&nbsp; **4)시그모이드 커널**:
$$
K(\mathbf{x}_i, \mathbf{x}_j) = \tanh(\kappa \mathbf{x}_i \cdot \mathbf{x}_j + \theta)
$$
여기서 κ와 𝜃는 상수입니다.

***KDE*** 

Kernel density estimation (KDE)은 확률 밀도 함수(Probability Density Function, PDF)를 추정하기 위한 비모수적 방법입니다. 데이터가 어떤 변수에서 어떻게 분포되어 있는지를 시각적으로 표현하는 데 사용됩니다.

여기서 주요한 개념은 확률 밀도 함수입니다. 확률 밀도 함수는 변수가 어떤 값을 가질 확률을 나타내는 함수로, 변수의 분포를 정의합니다. KDE는 이러한 확률 밀도 함수를 추정하는 방법 중 하나로, 주어진 데이터로부터 변수의 분포를 근사적으로 계산합니다.

KDE는 다음과 같은 과정을 통해 작동합니다:

데이터 포인트의 위치를 기준으로 밀도 측정: 주어진 데이터 포인트들을 중심으로 일정한 폭(대역폭, bandwidth)을 가지는 커널(kernel) 함수를 사용하여 각 데이터 포인트의 주변 밀도를 측정합니다.

모든 데이터 포인트에 대한 커널 함수 적용: 각 데이터 포인트에 대해 커널 함수를 적용하고, 이들을 모두 합하여 전체 데이터셋의 확률 밀도 함수를 추정합니다.

커널 함수의 대역폭 설정: KDE의 결과는 대역폭 선택에 따라 달라집니다. 대역폭이 작으면 추정된 확률 밀도 함수는 뾰족하고 데이터 포인트 주위에서 급격하게 변화합니다. 대역폭이 크면 보다 부드럽고 완만한 확률 밀도 추정이 이루어집니다.

따라서 Kernel Density Estimation (KDE) plot은 주어진 데이터의 변수(여기서는 'time'과 'age' 특성)의 분포를 부드럽게 나타내는 그래프입니다. 데이터 포인트의 밀도가 높은 지점은 KDE plot 상에서 더 높은 값으로 나타나며, 변수가 어떻게 분포되어 있는지 시각적으로 잘 이해할 수 있습니다.

### 전처리

```python
X=data_df.drop(["DEATH_EVENT"],axis=1)
y=data_df["DEATH_EVENT"]
```

```python
col_names = list(X.columns)
s_scaler = preprocessing.StandardScaler()
X_scaled = s_scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns = col_names)
X_scaled.describe().T
```

```python
#Plotting the scaled features using boxen plots
colors =["#CD5C5C","#F08080","#FA8072","#E9967A","#FFA07A"]
plt.figure(figsize=(20,10))
sns.boxenplot(data = X_scaled,palette = colors)
plt.xticks(rotation=60)
plt.show()
```

***Boxen Plot***

Seaborn의 boxenplot은 기본적으로 boxplot과 비슷하지만, 주로 데이터의 분포를 더 세부적으로 시각화하는 데 사용됩니다. 일반적인 boxplot은 주로 중앙값과 사분위수(1사분위수와 3사분위수)를 보여주며, 이상치를 표현할 수 있습니다. 반면에 boxenplot은 'letter-value' 플롯이라고도 불리며, 더 많은 사분위수를 포함하여 데이터의 분포를 상세하게 보여줍니다.

따라서 boxenplot은 여러 개의 박스를 생성하여 각 데이터 분포의 세부적인 정보를 제공하는 것이 아니라, 단일 박스 내에서 보다 많은 사분위수 정보를 제공하는 방식으로 설계되었습니다.

```python
#spliting variables into training and test sets
X_train, X_test, y_train,y_test = train_test_split(X_scaled,y,test_size=0.30,random_state=25)
```

### 모델 구축

```python
model1 = svm.SVC()
model1.fit(X_train, y_train)
y_pred = model1.predict(X_test)
model1.score(X_test, y_test)
```

```python
from sklearn.metrics import precision_score, recall_score, confusion_matrix, classification_report, accuracy_score, f1_score
```

```python
print(classification_report(y_test, y_pred))
```

***classification_report***

classification_report는 scikit-learn에서 제공하는 함수로, 분류 모델의 성능 평가를 종합적으로 보고하는 데 사용됩니다. 이 함수는 주어진 예측 결과와 실제 타겟 값(ground truth)을 기반으로 다음과 같은 여러 지표들을 포함하는 보고서를 생성합니다:

```python
# Getting the confusion matrix
cmap1 = sns.diverging_palette(2, 165, s=80, l=55, n=9)
plt.subplots(figsize=(10,7))
cf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(cf_matrix/np.sum(cf_matrix), cmap = cmap1, annot = True, annot_kws = {'size':25})
```
